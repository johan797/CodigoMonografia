Lidiar con un dataset de 75 millones de registros es un desafío significativo en términos de manejo de memoria y eficiencia. Vamos a abordar el problema paso a paso:

Dado el tamaño de tu dataset, es probable que no se  pueda cargar todo de una vez en memoria (dependiendo de la RAM de la máquina). Por lo tanto,se puede utilizar el parámetro chunksize para leer el dataset en chunks.

Queremos enfocarnos en los ataques DDoS. Por lo tanto, durante la lectura de cada chunk, filtra sólo los registros que corresponden a ataques DDoS y guarda estos registros filtrados

```python

# Rutas del archivo
archivo_entrada = 'dataset/data/nfuqnidsv2.csv'
archivo_salida = 'dataset/data/DDoS_attacks.csv'

# Tamaño de cada porción (chunk)
tamano_chunk = 1000000  # Este valor se puede ajustar según la memoria disponible
iterador_chunks = pd.read_csv(archivo_entrada, chunksize=tamano_chunk)

# Variable para saber si necesitamos escribir el encabezado al archivo de salida
escribir_encabezado = True

# Procesar y filtrar el dataset por porciones
for chunk in iterador_chunks:
    datos_filtrados = chunk[chunk['Attack'].isin(['DDoS', 'Benign'])]
    
    # Guardar los datos filtrados en el archivo de salida
    datos_filtrados.to_csv(archivo_salida, mode='a', header=escribir_encabezado, index=False)
    
    # Después del primer chunk, ya no necesitamos escribir el encabezado
    escribir_encabezado = False
```

Clasificaremos las variables en dos categorías principales: categóricas y numéricas. Por lo general, las direcciones y puertos IP son tratadas como categóricas aunque técnicamente son numéricas. Pero debido a su naturaleza y uso, se manejan comúnmente como categóricas. Las demás variables que se relacionan con conteos, tamaños, velocidades, y duraciones son consideradas numéricas.

1.Categóricas:

- IPv4 source address
- IPv4 destination address
- IPv4 source port number
- IPv4 destination port number
- IP protocol identifier byte
- Layer 7 protocol (numeric)
- ICMP Type * 256 + ICMP code
- ICMP Type
- DNS query transaction Id
- DNS query type (e.g. 1=A, 2=NS..)
- FTP client command return code

2. Numéricas:

- Incoming number of bytes
- Outgoing number of bytes
- Incoming number of packets
- Outgoing number of packets
- Flow duration in milliseconds
- Cumulative of all TCP flags
- Cumulative of all client TCP flags
- Cumulative of all server TCP flags
- Client to Server stream duration (msec) [Nota: Esta variable parece repetida]
- Min flow TTL
- Max flow TTL
- Longest packet (bytes) of the flow
- Shortest packet (bytes) of the flow
- Len of the smallest flow IP packet observed
- Len of the largest flow IP packet observed
- Src to dst Bytes/sec
- Dst to src Bytes/sec
- Number of retransmitted TCP flow bytes (src->dst)
- Number of retransmitted TCP flow packets (src->dst)
- Number of retransmitted TCP flow bytes (dst->src)
- Number of retransmitted TCP flow packets (dst->src)
- Src to dst average thpt (bps)
- Dst to src average thpt (bps)
- Packets whose IP size <= 128
- Packets whose IP size > 128 and <= 256
- Packets whose IP size > 256 and <= 512
- Packets whose IP size > 512 and <= 1024
- Packets whose IP size > 1024 and <= 1514
- Max TCP Window (src->dst)
- Max TCP Window (dst->src)
- TTL of the first A record (if any)
