Estas métricas son útiles para evaluar la calidad del modelo en términos de sus capacidades de clasificación. La matriz de confusión proporciona la información necesaria para calcular estas métricas. Además de estas métricas, se mencionan otras métricas y medidas como el coeficiente kappa de Cohen, el porcentaje de trazas recuperadas y el tiempo de ejecución total en el contexto de un modelo híbrido. Estas métricas pueden proporcionar información adicional sobre el rendimiento y la eficiencia del modelo en una tarea específica. El coeficiente kappa de Cohen, por ejemplo, mide la concordancia entre las predicciones del modelo y las etiquetas reales, y puede ayudar a cuantificar la calidad del modelo más allá de la probabilidad al azar. El porcentaje de trazas recuperadas puede ser relevante en aplicaciones específicas donde la recuperación de información es crítica. El tiempo de ejecución total es importante para evaluar la eficiencia del modelo en términos de recursos computacionales utilizados. En resumen, estas métricas y medidas permiten una evaluación completa y detallada del modelo híbrido en cuestión.

```python
Kappa mide el grado de concordancia de las evaluaciones nominales u ordinales realizadas por múltiples evaluadores cuando se evalúan las mismas muestras.

Por ejemplo, dos médicos diferentes examinan a 45 pacientes para determinar si tienen una enfermedad específica. ¿Con qué frecuencia coincidirá el diagnóstico de los médicos con respecto a la enfermedad (positivo o negativo)?
```

```python
from sklearn.metrics import cohen_kappa_score

# Ejemplo de etiquetas reales y predicciones
etiquetas_reales = [1, 2, 3, 1, 2, 3, 1, 2, 3]
predicciones = [1, 2, 3, 2, 2, 1, 1, 3, 3]

# Calcular el coeficiente Kappa de Cohen
kappa = cohen_kappa_score(etiquetas_reales, predicciones)

print("Coeficiente Kappa de Cohen:", kappa)
```

El Coeficiente Kappa de Cohen es una métrica diseñada originalmente para evaluar el acuerdo entre evaluadores humanos en problemas de clasificación o evaluación categórica. Su uso principal es medir la concordancia entre dos evaluadores humanos que deben tomar decisiones subjetivas sobre categorías.

En el contexto de modelos de aprendizaje profundo para clasificación binaria, el uso del Coeficiente Kappa de Cohen puede no ser la mejor elección, ya que no mide directamente el rendimiento del modelo, sino la concordancia entre las predicciones del modelo y las etiquetas reales. Además, el Coeficiente Kappa de Cohen se utiliza principalmente en situaciones en las que las categorías de clasificación son subjetivas y pueden no aplicarse directamente a problemas de clasificación binaria.

Para evaluar el rendimiento de un modelo de clasificación binaria, es más común utilizar métricas como la precisión, el recall, la F1-score, el área bajo la curva ROC (AUC-ROC), el área bajo la curva PR (AUC-PR), la matriz de confusión y otras métricas específicas para problemas de clasificación binaria. Estas métricas proporcionan información más directa sobre la capacidad del modelo para predecir correctamente las clases en un problema de clasificación binaria.

En resumen, aunque el Coeficiente Kappa de Cohen puede ser útil en contextos específicos de evaluación de acuerdo entre evaluadores humanos, no es la métrica más adecuada para evaluar el rendimiento de un modelo de aprendizaje profundo en un problema de clasificación binaria. Es preferible utilizar métricas diseñadas específicamente para este tipo de problemas.

### Comparacion modelo final

#### metricas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con los datos
data = {
    'Arquitectura': ['DNN', 'CNN', 'LSTM', 'GRU'],
    'Feature_Selection_Method': ['Mutual Information (MUTINFO)', 'Chi-squared (CHI2)', 'Entropy (EN)', 'Entropy (EN)'],
    'Exactitud': [83.4, 84.8, 86.8, 86.5],
    'F1_Score': [85.4, 86.3, 87.7, 87.4],
    'Precision': [76.2, 78.8, 82.4, 81.7],
    'Sensibilidad': [97.1, 95.3, 93.7, 94.1]
}

df = pd.DataFrame(data)

# Reorganizar los datos para que cada método de selección de características tenga su propia columna
df_long = pd.melt(df, id_vars=['Arquitectura', 'Feature_Selection_Method'], value_vars=['Exactitud', 'F1_Score', 'Precision', 'Sensibilidad'],
                  var_name='Metric', value_name='Value')

# Agregar el nombre de la arquitectura y el método de selección de características
df_long['Group'] = df_long['Arquitectura'] + '+' + df_long['Feature_Selection_Method'].str.split(' ').str[1]

# Crear el gráfico utilizando Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_long, x='Group', y='Value', hue='Metric')
plt.title('Métricas por arquitectura y método de selección de características')
plt.xlabel('Arquitectura y Método de Selección de Características')
plt.ylabel('Porcentaje %')
plt.xticks(rotation=45)
plt.legend(title='Métrica', loc='upper left', bbox_to_anchor=(1, 1))
plt.ylim(60, 100)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con los datos de tiempo
data = {
    'Arquitectura': ['DNN', 'CNN', 'DNN'],
    'Feature_Selection_Method': ['Mutual Information (MUTINFO)', 'Chi-squared (CHI2)', 'Eliminación Recursiva(RFE)'],
    'Tiempo': [136.94, 210.7, 1428.9],
}

df = pd.DataFrame(data)

# Combinar arquitectura y método de selección de características
df['Group'] = df['Arquitectura'] + '+' + df['Feature_Selection_Method'].str.split(' ').str[1]


# Crear el gráfico utilizando Seaborn

plt.figure(figsize=(5, 5))
sns.barplot(data=df, x='Group', y='Tiempo', hue='Group', palette='Set1', legend=False)
plt.title('Tiempo de entrenamiento Modelos comparados')
plt.xlabel('Arquitectura y Método de Selección de Características')
plt.ylabel('Tiempo (segundos)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
