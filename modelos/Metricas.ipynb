## Metricas

En ciberseguridad, un aspecto importante es la precisión de las herramientas y sistemas de seguridad para detectar amenazas y ataques. Para captar este concepto, nos referimos a cuatro términos: verdadero positivo, verdadero negativo, falso positivo y falso negativo.

1. Verdadero Positivo (TP): Identificar correctamente una amenaza
2. Verdadero Negativo (TN): Identificar correctamente que no hay amenaza
3. Falso positivo (FP): identificar erróneamente algo que no es una amenaza como una amenaza
4. Falso negativo (FN): no detectar una amenaza real


### Accuracy (Precisión):

Representa la proporción de predicciones correctas entre el número total de predicciones realizadas. Se calcula como: (Verdaderos Positivos + Verdaderos Negativos) / Total de Predicciones.

### Loss-rate:

Es el complemento de la precisión. Representa la proporción de predicciones incorrectas entre el número total de predicciones realizadas. Se calcula como: 1 - Accuracy.

### Precision (Precisión):
Indica cuán preciso es un modelo cuando predice la clase positiva. Se calcula como: Verdaderos Positivos / (Verdaderos Positivos + Falsos Positivos).

### False Positive Ratio:
Es el complemento de la Precision. Mide la proporción de falsos positivos entre el total de predicciones positivas. Se calcula como: Falsos Positivos / (Verdaderos Positivos + Falsos Positivos).

### Recall (Sensibilidad o Exhauscividad):
Mide la capacidad del modelo para identificar todos los ejemplos positivos reales. Se calcula como: Verdaderos Positivos / (Verdaderos Positivos + Falsos Negativos).

### True Positive Rate (TPR) o Sensitivity:
Es equivalente al Recall. Mide la proporción de verdaderos positivos entre el total de casos positivos reales.

### Detection Rate:
Es equivalente al Recall o TPR. Indica la proporción de positivos verdaderos entre el total de casos positivos reales.

### F1-score (Puntaje F1):

Es una métrica que combina Precision y Recall en un solo número. Es útil cuando se desea comparar dos modelos y no se tiene una preferencia clara entre Precision y Recall. Se calcula como: 2 * (Precision * Recall) / (Precision + Recall).

### False Negative Rate (FNR):

Es el complemento del Recall. Mide la proporción de falsos negativos entre el total de casos positivos reales. Se calcula como: Falsos Negativos / (Verdaderos Positivos + Falsos Negativos).

### AUC-ROC (Área Bajo la Curva - Característica Operativa del Receptor):

Representa el área bajo la curva ROC, que es un gráfico que compara el TPR (eje Y) con el FPR (False Positive Rate, eje X) para diferentes umbrales de decisión. Un AUC-ROC de 1 indica un modelo perfecto, mientras que un AUC-ROC de 0.5 indica un modelo no mejor que una elección aleatoria.

### AUC-PR (Área Bajo la Curva - Precision-Recall):

Es similar al AUC-ROC, pero utiliza un gráfico que compara la Precisión con el Recall para diferentes umbrales de decisión.

### Confusion Matrix (Matriz de Confusión):

Es una tabla que muestra las combinaciones de predicciones y valores reales, es decir, Verdaderos Positivos, Falsos Positivos, Verdaderos Negativos y Falsos Negativos.
