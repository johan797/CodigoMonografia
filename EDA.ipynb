### Conjunto de datos balanceado selección de las tres clases


```python
#'data2' es tu DataFrame y 'Attack' es la columna que contiene el tipo de ataque

# Filtrar y tomar una muestra de 1 millón de registros Benign
benign_sample = data2[data2['Attack'] == 'Benign'].sample(n=2000000, random_state=1)

# Filtrar y tomar una muestra de 1 millón de registros DDoS
ddos_sample = data2[data2['Attack'] == 'DDoS'].sample(n=2000000, random_state=1)

# Concatenar las muestras en un nuevo DataFrame
sampled_data = pd.concat([benign_sample, ddos_sample])

# Verificar el resultado
print(sampled_data['Attack'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un gráfico de conteo con colores diferentes para cada columna y ajustar el ancho de las barras
sns.set(style='whitegrid')  # Cambiar el estilo del fondo

# Definir una paleta de colores (aquí se usa la paleta 'Set3' con 12 colores)
custom_palette = sns.color_palette('Set1', n_colors=len(sampled_data['Attack'].unique()))

# Crear el gráfico de conteo con asignación de 'Attack' a 'hue' y desactivación de la leyenda
ax=sns.countplot(x='Attack', data=sampled_data, hue='Attack', palette=custom_palette, linewidth=2, legend=False)

# Cambiar nombres de ejes
ax.set_xlabel('Clase')  # Cambiar el nombre del eje x
ax.set_ylabel('Conteo')  # Cambiar el nombre del eje y

plt.title('Distribución de Tipos de Ataque en la Muestra')
plt.xticks(rotation=90)  # Girar las etiquetas del eje x si es necesario
plt.show()
```

### Transformación y codificación. Tratamiento columnas tipo object float y codificacion

```python
import pandas as pd
import numpy as np
df = pd.read_csv('dataset/data/sample_4M.csv')
```

```python
df.info()
```

```python
df.describe()
#description.to_excel("description.xlsx", engine='openpyxl')
```

```python
IPV4_SRC_ADDR:eliminar
IPV4_DST_ADDR:eliminar
L7_PROTO: redondeo al entero menor mas cercano
SRC_TO_DST_SECOND_BYTES:convertir a entero
DST_TO_SRC_SECOND_BYTES:convertir a entero
FTP_COMMAND_RET_CODE:convertir a entero
Attack:codificacion ordinal
Dataset:eliminar
label: eliminar
```

```python
# 1. Eliminar las columnas
df.drop(columns=['IPV4_SRC_ADDR', 'IPV4_DST_ADDR', 'Label','Dataset'], inplace=True)

# 2. Redondear la columna L7_PROTO al entero menor más cercano
df['L7_PROTO'] = df['L7_PROTO'].apply(lambda x: int(x))

# 3. Convertir las columnas especificadas a enteros
for column in ['SRC_TO_DST_SECOND_BYTES', 'DST_TO_SRC_SECOND_BYTES', 'FTP_COMMAND_RET_CODE']:
    df[column] = df[column].astype(int)

# 4.  Define el mapeo de codificación ordinal
attack_mapping = {
    "DDoS": 1,
    "Benign": 0,
}

# 5. Aplica la codificación directamente a la columna 'Attack'
df['AttackCod'] = df['Attack'].map(attack_mapping).astype(int)

# 6. elminacion columna Attack 
df.drop(columns=['Attack'], axis=1, inplace=True)
df.info()

tipos_ataques = df['AttackCod'].unique()

#  cuántas veces aparece cada tipo de ataque en el dataset`
conteo_ataques = df['AttackCod'].value_counts()
print(conteo_ataques)
```

```python
df.to_csv('dataset/data/sample_4M_cod.csv', index=False)
```
